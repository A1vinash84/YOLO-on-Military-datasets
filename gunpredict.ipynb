```python
from ultralytics import YOLO
import cv2
import math 
# start webcam
cap = cv2.VideoCapture(0)
cap.set(3, 640)
cap.set(4, 480)

# model
model = YOLO(r"C:\Users\AVINASH G\code\Minimilitia\military1\runs\detect\train6\weights\best.pt")

# object classes
classNames = ['Machinegun', 'Pistol', 'Rifle', 'Shotgun']


while True:
    success, img = cap.read()
    results = model(img, stream=True)

    # coordinates
    for r in results:
        boxes = r.boxes

        for box in boxes:
            # bounding box
            x1, y1, x2, y2 = box.xyxy[0]
            x1, y1, x2, y2 = int(x1), int(y1), int(x2), int(y2) # convert to int values

            # put box in cam
            cv2.rectangle(img, (x1, y1), (x2, y2), (255, 0, 255), 3)

            # confidence
            confidence = math.ceil((box.conf[0]*100))/100
            print("Confidence --->",confidence)

            # class name
            cls = int(box.cls[0])
            print("Class name -->", classNames[cls])

            # object details
            org = [x1, y1]
            font = cv2.FONT_HERSHEY_SIMPLEX
            fontScale = 1
            color = (255, 0, 0)
            thickness = 2

            cv2.putText(img, classNames[cls], org, font, fontScale, color, thickness)

    cv2.imshow('Webcam', img)
    if cv2.waitKey(1) == ord('q'):
        break

cap.release()
cv2.destroyAllWindows()
```

```python
import numpy as np
from ultralytics import YOLO
import cv2
import math
import itertools

# Start webcam
cap = cv2.VideoCapture(0)
cap.set(3, 640)
cap.set(4, 480)

# Load both YOLO models
model1 = YOLO(r"C:\Users\AVINASH G\code\Minimilitia\military1\runs\detect\train\weights\best.pt")
model2 = YOLO(r"C:\Users\AVINASH G\code\Minimilitia\military1\runs\detect\train6\weights\best.pt")

# Object classes (combined from both models)
classNames = ['Machinegun', 'Pistol', 'Rifle', 'Shotgun', 'Tank']  # Fill in all classes

while True:
    success, img = cap.read()

    # Perform inference with both models
    results1 = model1(img, stream=True)
    results2 = model2(img, stream=True)

    # Combine predictions (adjust based on your strategy)
    combined_results = itertools.chain(results1, results2)

    # Process combined predictions and perform segmentation
    for r in combined_results:
        boxes = r.boxes
        for box in boxes:
            x1, y1, x2, y2 = box.xyxy[0].int()  # Convert to integers

            # Create a mask for the detected object
            mask = np.zeros(img.shape[:2], dtype="uint8")
            mask[y1:y2, x1:x2] = 255

            # Apply segmentation (e.g., isolate object, apply effects)
            segmented_object = cv2.bitwise_and(img, img, mask=mask)

            # Bounding box (optional for visualization)
            
            cv2.rectangle(img, (int(x1), int(y1)), (int(x2), int(y2)), (255, 0, 255), 3)
# Ensure integer coordinates

            # Confidence
            confidence = math.ceil((box.conf[0] * 100)) / 100
            print("Confidence --->", confidence)

            # Class name
            cls = int(box.cls[0])
            print("Class name -->", classNames[cls])

            # Display segmentation results
            cv2.imshow('Segmented Object', segmented_object)
            org = (int(x1), int(y1))  # Ensure integer coordinates for org
            font = cv2.FONT_HERSHEY_SIMPLEX
            fontScale = 1
            color = (255, 0, 0)
            thickness = 2
            cv2.putText(img, classNames[cls], org, font, fontScale, color, thickness)


    cv2.imshow('Webcam', img)
    if cv2.waitKey(1) == ord('q'):
        break

cap.release()
cv2.destroyAllWindows()
```

```python
import cv2
from ultralytics import YOLO

# Load your trained YOLOv8 model
model = YOLO(r'C:\Users\AVINASH G\code\Minimilitia\military1\runs\detect\train\weights\best.pt')

# Open the video file
video_path = r"C:\Users\AVINASH G\Downloads\Shooting 54 Guns (TFBTV Compilation).mp4"  # Replace with your video file path
cap = cv2.VideoCapture(video_path)

# Loop through the video frames
while cap.isOpened():
    # Read a frame from the video
    ret, frame = cap.read()

    if ret:
        # Perform YOLOv8 inference on the frame
        results = model(frame)

        # Visualize the results on the frame
        annotated_frame = results[0].plot()
        
        # Display the annotated frame
        cv2.imshow("YOLOv8 Inference", annotated_frame)
        
        # Break the loop if 'q' is pressed
        if cv2.waitKey(1) & 0xFF == ord('q'):
            break
    else:
        # Break the loop if the end of the video is reached
        break

# Release the video capture object and close the display window
cap.release()
cv2.destroyAllWindows()
```
