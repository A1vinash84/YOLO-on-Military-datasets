```python
#Counting with both bounding boxes and labels (video)

import cv2
import numpy as np
from ultralytics import YOLO

# Load your trained YOLOv8 model
model = YOLO(r'C:\Users\AVINASH G\code\Minimilitia\military1\runs\detect\train\weights\best.pt')

# Open the video file
video_path = r"C:\Users\AVINASH G\Downloads\Shooting 54 Guns (TFBTV Compilation).mp4"  # Replace with your video path
cap = cv2.VideoCapture(video_path)

while True:
    # Capture frame-by-frame
    ret, frame = cap.read()

    if not ret:
        print("Can't receive frame (stream end?). Exiting...")
        break

    # Perform YOLOv8 inference on the frame
    results = model(frame)

    # Count the detected guns in the frame
    total_guns = 0
    if hasattr(results[0], 'pandas'):
        for _ in results[0].pandas().xyxy[0]:
            total_guns += 1
    else:
        for _ in results[0]:
            total_guns += 1

    # Draw the total gun count on the frame in blue
    font = cv2.FONT_HERSHEY_SIMPLEX
    text = f"Total Guns: {total_guns}"
    cv2.putText(frame, text, (10, 40), font, 0.8, (255, 0, 0), 2)

    # Access detections and visualize them manually
    boxes = results[0].boxes  # Access the bounding boxes
    for box in boxes:
        # Handle tensor-based coordinates and ensure integer values
        coordinates = box.xyxy[0].cpu().numpy()  # Move coordinates to CPU and convert to NumPy array
        x1, y1, x2, y2 = coordinates.astype(int).tolist()

        cv2.rectangle(frame, (x1, y1), (x2, y2), (0, 255, 0), 2)  # Draw the rectangle

        # Access and display the label for the detected object
        label = model.names[int(box.cls)]  # Access the label using model.names and the class index
        cv2.putText(frame, label, (x1, y2 - 5), font, 0.7, (0, 0, 255), 2)  # Draw the label above the rectangle

    # Display the resulting frame
    cv2.imshow("YOLOv8 Inference", frame)

    # Exit if the 'q' key is pressed
    if cv2.waitKey(1) == ord('q'):
        break

# Release video capture and close windows
cap.release()
cv2.destroyAllWindows()
```
